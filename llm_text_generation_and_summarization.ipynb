```python
from openai import OpenAI
import os
```

```python
import os
os.environ["Your_llm_model_API_KEY"] = "<Your_api_key>"
```

```python
api_key = os.getenv('Your_llm_model_API_KEY')
```

```python
client = OpenAI(base_url='https://Your_llm_model_API_KEY/api/', api_key=api_key)
```

```python
def generate_text(prompt):
    response = client.chat.completions.create(
        model="Your_llm_model_API_KEY",  
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        max_tokens=100,
        temperature=0.7
    )

    # Debug response
    print(response)

    content = response.choices[0].message.content

    if content:
        return content.strip()
    else:
        return "No response generated"
```

```python
prompt = "Once upon a time"
```

```python
generated_text = generate_text(prompt)
print(prompt, generated_text)
```

```python
def generate_text(prompt, max_tokens, temperature):
    response = client.chat.completions.create(
        model="Your_llm_model_API_KEY",  
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        max_tokens=max_tokens,
        temperature=temperature
    )

    content = response.choices[0].message.content

    if content:
        return content.strip()

    return "Model returned empty response"
```

```python
generated_text = generate_text(prompt, 50, 0)
print(prompt, generated_text)
```
